Classification - Sentiment Analysis

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("Instagram wants to limit hashtag spam.")
```

Zero-shot Classifier

```python
classifier = pipeline("zero-shot-classification", model = 'facebook/bart-large-mnli')

classifier(
    ["Inter Miami wins the MLS", "Match tonight betwee Chiefs vs. Patriots", "Michael Jordan plans to sell Charlotte Hornets"],
    candidate_labels=["soccer", "football", "basketball"]
    )
```

```python
new_lst = classifier(
    [" marked by India's exhilarating T20 World Cup victory over South Africa, ending an ICC trophy drought", "The season, which began in late 2023, concluded in March 2024. Puneri Paltan won their maiden title", \
     "Despite high expectations, the six-member contingent (four women, two men) returned without any medals, with top contenders like Nikhat Zareen, Lovlina Borgohain, and Amit Panghal exiting in early rounds"],
    candidate_labels = ["kabbadi", "boxing", "cricket", "other"])
```
